```python
import pandas as pd
import numpy as np
file1 = pd.read_csv('fifa21_train.csv')
```

```python
col = []
column_names = file1.columns
for i in column_names:
    col.append(i.lower())
file1.columns = col


```

```python
file1_cat = file1.select_dtypes('object')
file1_num = file1.select_dtypes('int')
file1_cat['hits']
```

```python
file1_cat['hits'] = pd.to_numeric(file1_cat['hits'], errors='coerce')
file1_cat['hits'].dtypes
```

```python
file1_cat['weight'] = file1_cat['weight'].map(lambda x: x.rstrip('lbs'))
file1_cat['weight'] = pd.to_numeric(file1_cat['weight'], errors='coerce')
file1_cat['weight']
```
